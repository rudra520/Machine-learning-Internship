Indentation is the 4-space or 1-tab gap in Python, which by default gets picked up from the new line whenever we end any previous Python statement with a colon.

# FOR Loop

```python
# syntax
for temp_var in iterable:
    body to the for loop
```

```python
list_1=[10,20,30,40]
for i in list_1:
    print(i)
```

```python
set_1={10,20,30,40,50,10}
for i in set_1:
    print(i)
```

```python
dict_1={"name":"IStudio","age":6,"speciality":"Internship"}

for i in dict_1 :
    print(i)
```

```python
for j in dict_1.values():
        print(j)
```

```python
for k in dict_1.items():
    print(k)
```

```python
# to find out even number  form first 10 nutural numbers.

for i in range (1,11):
    print(i)
```

```python
for i in range(2,12,2):
    print(i)
```

### The built-in **enumerate()** function in Python adds a automatic counter to any iterable object and returns it as an enumerate iterator yielding (index, item) pairs.

```python
list_1=[10,20,30,40]
for i in enumerate(list_1):
    print(i)
```

```python
list_2=[67,56,45,78]
list_1=[10,20,30,40]

for i,j in zip (list_1,list_2):
    print(i+j)
```
