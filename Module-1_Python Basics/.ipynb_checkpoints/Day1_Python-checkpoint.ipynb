## Keyword & Identifier

#### -Keywords in Python are special reserved words that define the structure and syntax of the language.

```python
#Example -
        # 1. Import the library - import is keyword
import numpy as np

# 2. Create a 1D array (Vector)
vector = np.array([1, 2, 3, 4, 5])
print("1D Array:", vector)
```

## Identifiers

1. We cannot use keywords as the identifiers.
2. Identifiers can start only with an alphabet or _.
3. Special characters are not allowed.

```python
#Example-
var_1=90
var_2=20
num= var_1 + var_2
print(num)

#here var_1 and var 2 is Identifiers.
```

```python
_hey=89 #i am declaring the identifier
```
