```python
# sum of two numbers using function:

def sum_1(num_1,num_2):
    return (num_1+num_2)
```

```python
# when to use print and when to use return:

def sub_1(a,b):
    return(a-b)
```

```python
def master(a1,b1):
    ans_1=sum_1(a1,b1)
    ans_2=sub_1(a1,b1)
    print(ans_1,ans_2)
```

```python
master(45,89)
```

```python
# variable length argument.
def sum_n(*param_1):
    sum_1=0
    for i in param_1:
        sum_1+=i
    print(sum_1)
```

```python
sum_n(9,9,0)
```

```python
def factorial(x):
    if x==1:
        return 1
    else:
        return x * factorial(x-1)
```
