# Tuple:

```python
tuple_1=(10,20,30.4,"Rudra",[10,20,30])
print(tuple_1)
```

```python
print(type(tuple_1))
```

```python
tuple_1[1:4]
```

```python
 tuple_1[::2]   #using jump or skip , becouse It is always going to happen sequentially so we can just jump or skip a thing which is not need or want to print.
```

```python
tuple_1[::-1]     # Reversing the tuple
```

```python
tuple_1[-1:-4:-1]    # Backward slicing
```
