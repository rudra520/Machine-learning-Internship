## **Array Attributes:**

```python
a1=np.arange(10).reshape(5,2) # arrange helps us to create same as that of the range function
                 # but  in 1D format
print(a1)
```

```python
a2=np.arange(12,dtype="float").reshape(3,4)
print(a2)
```

```python
a3=np.arange(11,dtype="complex")
print(a3)
```

```python
a2=a2.astype(np.int64)
```

```python
a2*4
```

```python
for i in a2:
    print(i)
```

```python
for i in np.nditer(a2):
    print(i)
```
