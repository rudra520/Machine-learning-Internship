# Operators: <small> Python operators are special symbols or keywords used to perform operations on values and variables. </small>

## Arithmetic Operators:

### +, -, /, *, **, %, //

```python
a=10
b=20
c=a+b
print(c)
```

```python
print(a**b)
```

```python
print(a//b)   # floor division--which divides two numbers and rounds the result down to the nearest whole number (towards negative infinity).
```

```python
print(a%b)  
```

```python
result = 10 % 3  # Output is 1 (3 goes into 10 three times, leaving 1)
print(result)
```
