---> lists follows indexing of both type forward indexing and backward indexing.

```python
list_1=[10,20,30,40,50,"Rudra",True,None]
print(list_1)
```

```python
print(list_1[5])     #forward indexing 
```

```python
print(list_1[-1])    #backward indexing
```
