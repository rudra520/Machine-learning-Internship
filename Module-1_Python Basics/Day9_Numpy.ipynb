# Stacking and Splitting of a matrix:

```python
mat_1=np.arange(12).reshape(3,4)
mat_2=np.arange(12, 24).reshape(3,4)
```

```python
mat_1
```

```python
mat_2
```

```python
v=np.vstack((mat_1, mat_2)) # it stack up the 2 matrix in vartical form.
v
```

```python
h=np.hstack((mat_1,mat_2)) # it stack up the 2 matrix in horizontal form.
h
```

```python
np.hsplit(v,2) # its split matrix in horizontal form on basis of columns.
```

```python
np.vsplit(h,3) # its split matrix in vertical form on basis of Row.
```
