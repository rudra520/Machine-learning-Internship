## Q3

```python
ar1=np.arange(0,9).reshape(3,3)
```

```python
ar2=np.arange(10,19).reshape(3,3)
```

```python
# 1 
#a - rowwise
np.vstack((ar1,ar2)) 
```

```python
# another way
print(np.concatenate([ar1,ar2],axis=0))
```

```python
#b - columnwise
np.hstack((ar1,ar2))
```

```python
# another way
print(np.concatenate([ar1,ar2],axis=1))
```
